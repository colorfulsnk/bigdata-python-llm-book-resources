### 3.1.2 路径分隔符

在指定文件路径时，Windows 与 macOS/Linux 的路径分隔符不同：

- **Windows**：使用反斜杠 `\`，例如：`C:\Users\Alice\Documents\file.txt`
- **macOS / Linux**：使用正斜杠 `/`，例如：`/home/alice/documents/file.txt`

在 Python 字符串中，反斜杠有转义含义，因此 Windows 路径需写成双反斜杠 `\\`，或使用原始字符串 `r"C:\Users\..."` 。

推荐使用 `pathlib.Path` 或 `os.path.join()` 处理路径，可自动适配不同操作系统：

```python
from pathlib import Path

# 方法一：直接使用完整路径（Mac/Linux）
# file_path = '/Users/XXX/Desktop/大数据分析材料/data/auto.csv'

# 方法二：先设定工作路径，再使用相对路径
import os
# os.chdir('/Users/XXX/Desktop/大数据分析材料/data')
# df = pd.read_csv('auto.csv')

# 方法三（推荐）：使用 Path 模块，自动处理斜杠
path = Path('data')          # 相对路径示例
file = path / 'auto.csv'    # 自动处理分隔符
print(file)                  # data/auto.csv (Mac/Linux) or data\auto.csv (Windows)
```

### 3.2.2 常用的正则表达式函数

```python
import re
pattern = "(\w+\.)*\w+@(\w+\.)+[A-Za-z]+"
#re.match（）函数
res=re.match(pattern,"123456789@qq.com")  #第一个参数为匹配规则，第二个参数为字符串
print("结果",res)   #注意：返回的是match对象Match object，而非直接匹配的结果
if res!=None:
    print("位置",res.span())  #返回匹配成功字符串的起始和结束位置
```

```python
pattern="(\w+\.)*\w+@(\w+\.)+[A-Za-z]+" 
res=re.match(pattern,"我的邮件地址:123456789@qq.com") 
print("结果",res) 
if res!=None:
    print("位置",res.span())
```

```python
re.search(r"\d+", "课程编号是：101；分数是：99")
```

```python
pattern="(\w+\.)*\w+@(\w+\.)+[A-Za-z]+" 
mystr="我的邮件地址:123456789@qq.com"
res=re.search(pattern,mystr)
print ("结果",res)   #成功匹配
if res!=None:
    print("位置：",res.span())
    print("匹配的字符串：",res.group())
```

```python
re.findall(r"\d+", "课程编号是：101；分数是：99")
```

```python
pattern="(\w+\.)*\w+@(\w+\.)+[A-Za-z]+" 
mystr3="我的邮件地址:123456789@qq.com；以及987654321@gmail.com" 
res=re.findall(pattern,mystr3)
print ("结果",res)
```

```python
import re
pattern = "(?:\w+\.)*\w+@(?:\w+\.)+[A-Za-z]+" #每个括号中在最前面加上"?:",告诉python这不是捕获组
mystr3="我的邮件地址:123456789@qq.com；987654321@gmail.com" 
res = re.findall(pattern, mystr3)
print("结果", res)
```

```python
pattern="(\w+\.)*\w+@(\w+\.)+[A-Za-z]+" 
mystr="我的邮件地址:123456789@qq.com;"
res=re.sub(pattern,"987654321@gmail.com",mystr) #第一个参数为正则表达式，第二个参数为替换后的字符，第三个参数为原始数据串
print(mystr)
print(res)
```

```python
print("承担诉讼费一方：", fee_bearer[0])
```

```python
if fee_bearer[0] == "原告":
    print("原告承担费用 → 原告败诉")
else:
    print("被告承担费用 → 被告败诉")
```

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm


# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)   #生成网格矩阵
R = np.sqrt(X**2 + Y**2)   #计算每个点到原点的距离
Z = np.sin(R)      #计算距离为 R 的点的正弦值，形成 Z 值

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)  #在 3D 轴上绘制三维曲面图，指定了最小值和颜色映射。
plt.savefig("img/3d_plot.pdf", format='pdf', bbox_inches='tight') 
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

### 3.5.4 绘图进阶应用

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = 1/np.tan(x)  #numpy中没有cot函数
# 创建子图
#创建了一个包含 4 个子图的图形对象 fig 和一个包含这些子图的轴对象列表 axs。每个子图都可以通过 axs 列表中的索引进行访问和操作
fig, axs = plt.subplots(4)
# 绘制第一个子图
axs[0].plot(x, y1)
axs[0].set_title('Sin')
# 绘制第二个子图
axs[1].plot(x, y2)
axs[1].set_title('Cos')
# 绘制第三个子图
axs[2].plot(x, y3)
axs[2].set_title('Tan')
# 绘制第四个子图
axs[3].plot(x, y4)
axs[3].set_title('Cot')
# 调整布局
plt.tight_layout()
# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cot(x):
    if np.isclose(x % np.pi, 0).any():
        return None
    else:
        return 1 / np.tan(x)
x = np.linspace(0.01, 10, 100)  # 为避免 x % np.pi 出现零值，将初始值设为 0.01
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = cot(x)  # 引用我们定义的cot函数
# 创建子图 2*2布局
fig, axs = plt.subplots(2,2) 

axs[0,0].plot(x, y1)
axs[0,0].set_title('Sin')
axs[0,1].plot(x, y2)
axs[0,1].set_title('Cos')
axs[1,0].plot(x, y3)
axs[1,0].set_title('Tan')
axs[1,1].plot(x, y4)
axs[1,1].set_title('Cot')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager
from scipy.stats import norm

# ===== 设置中文字体路径 =====
font_path = 'font/Songti.ttc'
my_font = font_manager.FontProperties(fname=font_path)

# 避免负号乱码
plt.rcParams['axes.unicode_minus'] = False


def draw_custom_style_demo(style='default'):
    # 应用主题样式
    plt.style.use(style)

    # 设置随机种子（保证图形一致）
    np.random.seed(42)

    # 创建图形与2×2子图
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

    # ---------------- 子图1：正态分布密度 ----------------
    x = np.linspace(-5, 5, 200)
    for mu, sigma in [(-1, 0.8), (0, 1), (1.5, 0.5)]:
        y = norm.pdf(x, loc=mu, scale=sigma)
        axes[0, 0].plot(x, y, label=f'μ={mu}, σ={sigma}', linewidth=2)
    axes[0, 0].set_title("正态分布密度函数", fontproperties=my_font)
    axes[0, 0].legend(prop=my_font)

    # ---------------- 子图2：指数分布直方图 ----------------
    exp_data = np.random.exponential(scale=1.0, size=500)
    axes[0, 1].hist(exp_data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    axes[0, 1].set_title("指数分布直方图", fontproperties=my_font)

    # ---------------- 子图3：三组数据箱线图 ----------------
    data_box = [np.random.normal(loc=mu, scale=1.0, size=100) for mu in [0, 1, 2]]
    axes[1, 0].boxplot(data_box, labels=['A', 'B', 'C'])
    axes[1, 0].set_title("分组箱线图", fontproperties=my_font)

    # ---------------- 子图4：面积图 ----------------
    t = np.arange(0, 10, 0.1)
    y1 = np.sin(t) + 0.5 * np.random.randn(len(t))
    y2 = np.cos(t) + 0.5 * np.random.randn(len(t))
    axes[1, 1].fill_between(t, y1, alpha=0.5, label='系列1')
    axes[1, 1].fill_between(t, y2, alpha=0.5, label='系列2')
    axes[1, 1].set_title("面积图（趋势模拟）", fontproperties=my_font)
    axes[1, 1].legend(prop=my_font)

    # ---------------- 总标题 ----------------
    style_name = style.split('-')[0]  # 提取样式前缀
    plt.suptitle(f"Matplotlib 风格演示：{style_name}", fontsize=14, fontproperties=my_font)

    # 自动调整子图间距，避免遮挡
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.savefig(f'style_demo_{style_name}.pdf', format='pdf', bbox_inches='tight')

    # 显示图像
    plt.show()
```

### 3.6.2代码改进

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
from matplotlib.font_manager import FontProperties

# 设置中文字体路径
font_path = "font/Songti.ttc"
zhfont = FontProperties(fname=font_path)

# 设置图形风格
plt.style.use('seaborn-v0_8')

def plot_path(ax, αs, s_vals, δs, T=50):
    """
    在指定坐标轴上绘制所有参数组合下的 Solow 模型资本路径。
    参数:
        ax: 子图对象
        αs, s_vals, δs: 要枚举的参数列表
        T: 模拟期数
    """
    k = np.empty(T)
    for (α, s, δ) in product(αs, s_vals, δs):
        k[0] = 1
        for t in range(1, T):
            k[t] = s * k[t-1]**α + (1 - δ) * k[t-1]
        label = f"α = {α}, 储蓄率 = {s}, 折旧率 = {δ}"
        ax.plot(k, label=label)

    ax.set_xlabel("时间", fontproperties=zhfont)
    ax.set_ylabel("人均资本", fontproperties=zhfont)
    ax.set_ylim(0, 18)
    ax.grid(True, lw=0.3)
    ax.legend(loc='upper left', prop=zhfont, fontsize=9)

# 参数组合设定
param_sets = [
    ([0.25, 0.33, 0.45], [0.35], [0.08]),          # 变化 α
    ([0.33], [0.3, 0.4, 0.5], [0.08]),             # 变化 s
    ([0.33], [0.35], [0.05, 0.08, 0.12])           # 变化 δ
]

titles = ["变化 α", "变化 储蓄率 s", "变化 折旧率 δ"]

# 创建图形
fig, axes = plt.subplots(3, 1, figsize=(9, 14))

for ax, params, title in zip(axes, param_sets, titles):
    plot_path(ax, *params)
    ax.set_title(f"参数敏感性分析：{title}", fontproperties=zhfont)

plt.tight_layout()
plt.show()
```
