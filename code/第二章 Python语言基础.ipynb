```python
# 比较运算符示例
a, b = 10, 5
print(a > b)   # True
print(a < b)   # False
print(a == b)  # False
print(a != b)  # True
print(a >= 10) # True
print(b <= 3)  # False

# 成员运算符
lst = [1, 2, 3]
print(2 in lst)      # True
print(5 not in lst)  # True

# 逻辑运算符
x = 7
print(x > 3 and x < 10)  # True
print(x < 3 or x > 5)   # True
print(not x == 7)        # False
```

```python
import time

# 方法一：使用局部变量
def sum_with_local():
    total = 0
    for i in range(100000000):
        total += i
    return total

# 方法二：使用全局变量
total_global = 0
def sum_with_global():
    global total_global
    for i in range(100000000):
        total_global += i
    return total_global

# 测试局部变量性能
start = time.time()
sum_with_local()
end = time.time()
print(f"使用局部变量耗时：{end - start:.6f} 秒")

# 测试全局变量性能
# 注意：重新初始化全局变量
total_global = 0
start = time.time()
sum_with_global()
end = time.time()
print(f"使用全局变量耗时：{end - start:.6f} 秒")
```
